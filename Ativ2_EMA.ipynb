Importação de bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

Parametrização

```python
ε = 1.5  #Força eletromotriz em Volts
r = 1.0  #Resistência interna em Ohms
```

Cálculo das potências

```python
#Vetor para os valores de R (0 < R < 10)
Valores_R = np.linspace(0.1, 10, 100)

#Potência dissipada para cada instância R
Valores_P = (4 * ε**2 * Valores_R) / ((Valores_R + 2 * r) ** 2)
```

Plotagem do gráfico

```python
plt.figure(figsize=(8, 5))
plt.plot(Valores_R, Valores_P, label="Potência dissipada", color='b')
plt.axvline(2*r, linestyle="--", color="r", label="R = r (Máxima potência)")
plt.xlim([0, 10])
plt.xlabel("Resistência R (Ω)")
plt.ylabel("Potência dissipada (W)")
plt.title("Potência dissipada no resistor em função de R")
plt.legend()
plt.grid()
plt.show()
```
