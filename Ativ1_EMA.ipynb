Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

Definição dos parâmetros teóricos

```python
e0 = 9.1 #Tensão da bateria em volts
R = 9.87e3 #Resistência em Ohms
C = 1e-3 #Capacitância em Faradays
tau = R*C #Constante de tempo
t = np.linspace(0, 6 * tau, 1000) #Vetor de tempo

#Definição da equação de Vc(t)
Vc_t = e0*(1-np.exp(-t/tau))
```

Dados obtidos experimentalmente (Vídeo)

```python
Vc_t_real = [0.37, 1.36, 1.96, 2.76, 3.23, 3.67, 4.27, 4.63, 4.96, 5.26, 5.79,
             6.03, 6.25, 6.46, 6.73, 6.9, 7.05, 7.2, 7.39, 7.5, 7.61,
             7.8, 7.89, 8.01, 8.08, 8.14, 8.23, 8.29, 8.36, 8.41, 8.45,
             8.51, 8.54, 8.57, 8.62, 8.64, 8.68, 8.7, 8.72, 8.74, 8.76,
             8.79, 8.81, 8.82, 8.84, 8.85, 8.86, 8.87, 8.89, 8.89, 8.9,
             8.91, 8.92, 8.93, 8.93, 8.94, 8.94, 8.95, 8.95, 8.95, 8.96] #Valores retirados de cada segundo do vídeo (0~60)

#Criação de um vetor de tempo para os dados experimentais
t_exp = np.arange(len(Vc_t_real))
```

Plot e exibição do gráfico

```python
plt.figure(figsize=(8, 5))
plt.plot(t, Vc_t, label=r'$V_c(t)$ teórico', color='b')
plt.plot(t_exp, Vc_t_real, label=r'$V_c(t)$ experimental', color='g', linestyle='-.')
plt.axhline(y = 9.1, color='r', linestyle='--', label='V da bateria (9.1V)')
plt.xlim([0, 60])
plt.xlabel('Tempo (s)')
plt.ylabel('Tensão no Capacitor $V_C$ (V)')
plt.title('Carga Teórica e Experimental de um Capacitor')
plt.legend()
plt.grid()

plt.show()
```
