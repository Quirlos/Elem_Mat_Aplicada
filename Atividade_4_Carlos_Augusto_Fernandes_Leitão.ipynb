Importação de biblioteca

```python
import numpy as np
import matplotlib.pyplot as plt
```

Parametrização

```python
#Definição dos dados
m = 3.0 #Massa do bloco (Kg)
dt = 1 #Salto de tempo Δt (s)

t_values = np.arange(12) #Vetor de intervalo de tempo (0 a 11s)
F_x_values = np.array([-3, 6, 6, 6, 6, 6, 0, -4, -4, -4, -4, 0]) #Vetor de força nos tempos t
```

Cálculos de posição e velocidade

```python
a_values = F_x_values / m #Cálculo da aceleração

#Cálculo da velocidade e posição
v_x_values = np.zeros_like(t_values, dtype=float) #Vetor com os valores de velocidade
x_values = np.zeros_like(t_values, dtype=float) #Vetor com os valores de posição
v_x_values[0] = -3.0 #Condição inicial de velocidade

for i in range(1, len(t_values)):
    v_x_values[i] = v_x_values[i - 1] + a_values[i - 1] * dt
    x_values[i] = x_values[i - 1] + v_x_values[i - 1] * dt
```

Print do gráfico de movimento

```python
#Criar gráfico
plt.figure(figsize=(10, 5))
plt.plot(t_values, x_values, marker='o', linestyle='-', label="Posição x(t)")
plt.xlabel("Tempo (s)")
plt.ylabel("Posição do bloco (m)")
plt.title("Gráfico da Posição x(t) do bloco")
plt.grid()
plt.legend()
plt.show()
```
