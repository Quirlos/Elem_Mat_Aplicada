Importação de bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

Definição dos vetores dos dados de decaimento

```python
rad_sem_amostra = np.array([14, 21, 25, 20, 15, 21, 26, 22, 30, 19, 22, 18, 15, 18, 20,
                        22, 27, 21, 20, 28, 22, 17, 13, 16, 13, 20, 19, 20, 15, 27,
                        19, 24, 12, 22, 21, 23, 22, 20, 29, 32, 18, 20, 21, 28]) #Dados de radiação de fundo sem amostra

rad_com_amostra = np.array([127, 119, 130, 159, 125, 134, 105, 128, 135, 122, 122, 123,
                            134, 114, 141, 134, 123, 125, 114, 120, 119, 135, 127, 124,
                            144, 123, 117, 132, 137, 132, 135, 158, 125, 122, 126, 123,
                            123, 132, 119, 132, 106, 119, 123, 145, 129]) #Dados de radiação de fundo com amostra
```

Plotagem dos histogramas

```python
plt.hist(rad_sem_amostra)
plt.xlabel("Distribuição da radiação de fundo")
plt.ylabel("Frequência")
plt.title("Histograma - Radiação de fundo (sem amostra)")
plt.show()

plt.hist(rad_com_amostra, color="red")
plt.xlabel("Distribuição da radiação da amostra")
plt.ylabel("Frequência")
plt.title("Histograma - Radiação de amostra")
plt.show()
```
